```python
from nltk.tokenize import word_tokenize
from nltk import FreqDist
```

```python
tokens= word_tokenize(text)

# FreqDist(): A frequency distribution for the outcomes of an experiment. A frequency distribution records the number of times each outcome of an experiment has occurred.
freq = FreqDist(tokens)
freq.plot(20)
```

```python
tokens
```

```python
from nltk.corpus import stopwords

swords = stopwords.words('english')

words = [word.lower() for word in tokens if word.lower() not in swords and word.isalnum()]
print(words)
```

```python
freq = FreqDist(words)
freq.plot(20)
```

```python
freq.most_common(10)
```

```python
# Get the Link -> Read the link -> Beautify the text -> Tokenize the Text -> Remove the Stopwords -> Count the Frequencies of Word.

url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
data = url.read()

soup = BeautifulSoup(data, 'html.parser')

soup.find('title')

text = soup.get_text()
print(text)

tokens= word_tokenize(text)

swords = stopwords.words('english')

words = [word.lower() for word in tokens if word.lower() not in swords and word.isalnum()]
print(words)

freq = FreqDist(words)
freq.plot(20)

freq.most_common(10)
```

```python
# Get the Link -> Read the link -> Beautify the text -> Tokenize the Text -> Remove the Stopwords -> Lemmatize it -> Count the Frequencies of Word.

from nltk.stem import PorterStemmer, WordNetLemmatizer

url = urlopen('https://en.wikipedia.org/wiki/Tiger')
data = url.read()

soup = BeautifulSoup(data, 'html.parser')

soup.find('title')

text = soup.get_text()
print(text)

tokens= word_tokenize(text)

swords = stopwords.words('english')

ps = PorterStemmer()
wnl = WordNetLemmatizer()

words = [wnl.lemmatize(word.lower(), pos='n') for word in tokens if word.lower() not in swords and word.isalnum()]
print(words)

freq = FreqDist(words)
freq.plot(20)

freq.most_common(10)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

url = urlopen('https://mr.wikipedia.org/wiki/%E0%A4%B5%E0%A4%BE%E0%A4%98')
data = url.read()

soup = BeautifulSoup(data, 'html.parser')

soup.find('title')

text = soup.get_text()
print(text)

tokens= word_tokenize(text)

swords = stopwords.words('marathi')

words = [word for word in tokens if word.lower() not in swords and word.isalnum()]
print(words)

freq = FreqDist(words)
freq.plot(20)

freq.most_common(10)
```
