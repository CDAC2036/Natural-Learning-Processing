### Data Preparation

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
sent = 'Hello friends! How are you? I like Python Programming'
```

```python
# 1. Tokenization

tokens1 = word_tokenize(sent)
tokens1
```

```python
# 2. Remove puctuation

tokens2 = [token for token in tokens1 if token.isalnum()]
tokens2
```

```python
# 3. Remove Stopwords

swords = stopwords.words('english')
tokens3 = [word for word in word_tokenize(sent) if word.lower() not in swords and word.isalnum()]
tokens3
```

```python
# 4. Stemming

ps = PorterStemmer()

tokens4=[ps.stem(word) for word in tokens3]
tokens4
```

```python
def clean_text(sent):
    tokens = [ps.stem(word) for word in word_tokenize(sent) if word.lower() not in swords and word.isalnum()]
    return tokens
```

```python
X.apply(clean_text)
```

### TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
help(TfidfVectorizer)
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text)

X_new = tfidf.fit_transform(X)
```

```python
X_new
# 6854 are features
```

```python
X_new.data
```

```python
import pandas as pd

df = pd.DataFrame(columns=tfidf.get_feature_names_out(), data=X_new.toarray())
df
```

### Cross Validation


```python
from sklearn.model_selection import train_test_split
```

```python
# Default split is 75:25
X_train, X_test, y_train, y_test = train_test_split(X_new, y, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Model Building

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='linear')
```

```python
# %%time - for CPU time (works in jupyter) else import time
svm.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix, classification_report

y_pred = svm.predict(X_test)

print('Accuracy Score: ',accuracy_score(y_test,y_pred))
print('\nConfusion Matrix:\n',confusion_matrix(y_test,y_pred))
print('\nClassification Report:\n',classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
