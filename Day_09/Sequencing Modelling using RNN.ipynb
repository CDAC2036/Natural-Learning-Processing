```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
idx2char = dict(enumerate(alphabets))
idx2char
```

```python
char2idx = {v:i for i, v in enumerate(alphabets)}
char2idx
```

### Data Preparation

```python
seq_length = 1
X = [] # X would be from A to Y
y = [] # Y would be from B to Z
for i in range(len(alphabets) - seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    X += [[char2idx[seq_in[0]]]] # List of List
    y += [char2idx[seq_out]] # List
    print(seq_in, '->', seq_out)
```

```python
X
```

```python
y
```

### Prepare the Output Variable

```python
from keras.utils import to_categorical
```

```python
# to_categorical(y): Converts a class vector (integers) to binary class matrix.
y_new = to_categorical(y) # As there is no '0' hence 0th index will be 0 itself
y_new
```

### Prepare the Input Variable


```python
import numpy as np
```

```python
X = np.reshape(X, (25, 1, 1)) # Or we can do X.reshape(25,1,1)
X
```

```python
X.shape
```

```python
# Normalizing the Input data

X_new = X / 25 # As we have 25 alphabets in X, Hence we are dividing it by 25
```

```python
X_new
```

```python
X_new.shape
```

```python
y_new.shape
```

### Model Building

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation = 'relu'))
model.add(LSTM(32)) # By default activation is 'tanh'
model.add(Dense(26, activation='softmax')) # y_new.shape[1] = 26
```

```python
# Compile the Model

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_new, y_new, epochs = 1000, batch_size = 1)
```

```python
from keras.utils import plot_model

plot_model(model, show_shapes = True, show_layer_names = True, show_layer_activations=True, to_file = 'model.png')
```

```python
model.evaluate(X_new, y_new, batch_size = 1)
```

### Predict The character in Sequence

```python
pred = input('Enter the Character: ')

# Predict the character in sequence

input_eval = [char2idx[pred]]
input_eval = np.reshape(input_eval, (1, 1, 1))
input_eval = input_eval / 25

output = model.predict(input_eval)
print(idx2char[output.argmax()])
```

```python
seq_length = 3
X = [] # X would be from A to Y
y = [] # Y would be from B to Z
for i in range(len(alphabets) - seq_length):

    seq_in = alphabets[i:i + seq_length]
    seq_out = alphabets[i + seq_length]
    X.append([char2idx[char] for char in seq_in])
    y.append(char2idx[seq_out])
    print(seq_in, '->', seq_out)
```

```python
X
```

```python
y
```

```python
# Normalizing the input

X = np.reshape(X, (len(X), seq_length, 1))
X = X / 25
X
```

```python
X.shape
```

```python
y = to_categorical(y)
y
```

```python
y.shape
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(3,1), activation = 'relu'))
model.add(LSTM(128)) # By default activation is 'tanh'
model.add(Dense(26, activation='softmax')) # y_new.shape[1] = 26
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X, y, epochs = 1000, batch_size = 1)
```

```python
from keras.utils import plot_model

plot_model(model, show_shapes = True, show_layer_names = True, show_layer_activations=True, to_file = 'model.png')
```

```python
model.evaluate(X, y, batch_size = 1)
```
