```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertForSequenceClassification
```

```python
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = TFBertForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

```python
# Define the text we want

text = ["This was an amazing product.", "This is the worst thing I have ever bought."]
```

```python
# Tokenize the text

inputs = tokenizer(text, return_tensors='tf', padding=True, truncation=True, max_length = 30)
```

```python
# Get the model's prediction

output = model(inputs)
predictions = tf.argmax(output.logits, axis=1)
```

```python
labels = ['Negative', 'Positive']
for text, prediction in zip(text, predictions):
    print(f"{text}: {labels[prediction]}")
```
