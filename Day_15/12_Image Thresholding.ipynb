```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../images/Lenna.png',0)

binary_image = image > 127 # False(0/0) is black and True(1/255) is white (based on original Image)

plt.figure(figsize=(10,6))

plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(122), plt.imshow(binary_image, cmap='gray'), plt.title('Image Thresholding')
plt.show()
```

```python
binary_image
```

```python
binary_image = binary_image.astype(np.uint8) * 255 # Grayscaling
binary_image
# Then we can use cv2.imshow()
# cv2.imshow('Original Image', image)
# cv2.imshow('Image Thresholding', binary_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
